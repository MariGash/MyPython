# Условие 2: 2 задача
## 2.1. Изучите распределение домов от наличия вида на набережную
Постройте график\
Сделайте выводы

```python
memory_1 = df["waterfront"].value_counts()
memory_1
```

```python
names = memory_1.index
values = memory_1.values
```

```python
plt.figure(figsize=(10, 7))
explode = [0.5, 0.7]
plt.pie(values, autopct='%1.1f%%', labels = ["Без вида на набережную", "С видом на набережную"], explode=explode, colors = ["0.5", "#FF00BB",])
plt.title('Доля домов в зависимости от наличия вида на набережную')
plt.legend(loc='best');
```

Домов с видом на набережную менее 1%.

## 2.2. Изучите распределение этажей домов

```python
memory_2 = df["floors"].value_counts()
memory_2
```

```python
names = memory_2.index
values = memory_2.values
```

```python
plt.figure(figsize=(10, 8))
expl = [0.02, 0.02, 0.03, 0.04, 0.05, 0.2]
plt.pie(values, autopct='%1.1f%%', labels=names, explode=expl)
plt.title('Доля домов в зависимоти от этажности')
plt.legend(names, loc='upper right');
```

Почти половина домов от всего количества недвижимоти одноэтажные (49,4%). Домов в 2 этажа 38,1%. Самая маленькая доля домов в 3,5 этажа (менее 1%).

### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_lot'], kind='reg');
```

Из приведенных выше графиков видно, что стоимость домов в прямой зависимости от размеров жилой площади. 

```python
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg');
```

```python
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Наличие (1) / отсутствие (0) вида')
plt.title('Анализ в зависимости от вида на набережную');
```

Дома с видом на набережную имеют более высокую стоимость.

```python
sns.boxplot(x=df['price'], y=df['condition'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Оценка состояния')
plt.title('Анализ в зависимости от оценки состояния дома');
```

Самая высокая стоимость у домов с оценкой отлично. У домов с оценкой хорошо и удовлетворительно наблюдается большое количество выбросов значительно выше средней стоимости. Рассмотрим подробнее дома со стоимостью более 4 млн. руб.

```python
price_1 = df[df['price'] > 4000000]
price_1
```

```python
plt.figure(figsize=(6, 4))

sns.histplot(price_1['grade'], bins=20)

plt.title('Анализ домов со стоимостью выше 4 млн. руб.')
plt.xlabel('Качество строительства и дизайна')
plt.ylabel('Количество домов');
```

У домов с оценкой состояния хорошо и удовлетворительно со стоимостью выше 4 млн. руб. наблюдается высокий уровень качества строительства и дизайна (от 11 баллов).

```python
plt.figure(figsize=(6, 4))

sns.histplot(price_1['floors'], bins=20)

plt.title('Анализ домов со стоимостью выше 4 млн. руб.')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов');
```

Дома с оценкой состояния хорошо и удовлетворительно со стоимостью выше 4 млн. руб. имеют в основном 2 этажа.
